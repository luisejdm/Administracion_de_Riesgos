```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns
%matplotlib inline
```

- Gráfica de la distribución de los datos
- Medidas de tendencia central (Media y Mediana)
- Medidas de dispersión (Varianza y Desviación Estándar)
- Calcula y concluye sobre el sesgo y la curtosis de los datos.
- Grafica e interpreta un boxplot.


```python
prices = yf.download(['PLTR', 'MELI'], start='2020-01-01')['Adj Close'].dropna()
prices.head()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(prices, label=prices.columns, bins=10)
plt.legend(loc='best')
plt.show()
```

```python
rends = prices.pct_change().dropna()
plt.figure(figsize=(12, 6))
plt.hist(rends, label=rends.columns, bins=50, alpha=0.5)
plt.legend(loc='best')
plt.show()
```

```python
rends.mean()*252*100
```

```python
rends.std()*np.sqrt(252)*100
```

```python
rends.skew()
```

```python
rends.kurt()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(rends)
```
